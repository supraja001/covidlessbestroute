```python

from collections import defaultdict, deque


class Graph(object):
    def __init__(self):
        self.nodes = set()
        self.edges = defaultdict(list)
        self.distances = {}

    def add_node(self, value):
        self.nodes.add(value)

    def add_edge(self, from_node, to_node, distance1,distance2):
        self.edges[from_node].append(to_node)
        self.edges[to_node].append(from_node)
        self.distances[(from_node, to_node)] = distance1
        self.distances[(to_node, from_node)] = distance2


def dijkstra(graph, initial):
    visited = {initial: 0}
    path = {}

    nodes = set(graph.nodes)

    while nodes:
        min_node = None
        for node in nodes:
            if node in visited:
                if min_node is None:
                    min_node = node
                elif visited[node] < visited[min_node]:
                    min_node = node
        if min_node is None:
            break

        nodes.remove(min_node)
        current_weight = visited[min_node]

        for edge in graph.edges[min_node]:
            try:
                weight = current_weight + graph.distances[(min_node, edge)]
            except:
                continue
            if edge not in visited or weight < visited[edge]:
                visited[edge] = weight
                path[edge] = min_node

    return visited, path


def shortest_path(graph, origin, destination):
    visited, paths = dijkstra(graph, origin)
    full_path = deque()
    _destination = paths[destination]

    while _destination != origin:
        full_path.appendleft(_destination)
        _destination = paths[_destination]

    full_path.appendleft(origin)
    full_path.append(destination)
    #return list(full_path)
    return visited[destination], list(full_path)

if __name__ == '__main__':
    graph = Graph()
    nodes=['ADILABAD','NIRMAL','KOMARAMBHEEM ASIFABAD','NIZAMABAD','JAGITYAL',     'MANCHERIAL','KAMAREDDY','SIRCILLA','PEDDAPALLI','KARIMNAGAR','SANGAREDDY','MEDAK',
    'SIDDIPET','MEDCHAL','WARANGAL','JANGAON','YADADRI BHONGIR','MULUGU','JAYSHANKAR BHUPALPALLY','MAHABUBNAGAR','MAHABUBABAD','VIKARABAD','GHMC','NALGONDA','SURYAPET',
    'KOTHAGUDEM','NARAYANPET','NAGAR KURNOOL','KHAMMAM','WANAPARTHY','GADWAL','RANGAREDDY']

    for node in nodes:
        graph.add_node(node)
    
    graph.add_edge('ADILABAD','NIRMAL',dict['ADILABAD'],dict['NIRMAL'])
    graph.add_edge('ADILABAD','ASIFABAD',dict['ADILABAD'],dict['ASIFABAD']) 
    graph.add_edge('ASIFABAD','MANCHERIAL',dict['ASIFABAD'],dict['MANCHERIAL'])
    graph.add_edge('NIRMAL','NIZAMABAD',dict['NIRMAL'],dict['NIZAMABAD'])
    graph.add_edge('NIRMAL','JAGITYAL',dict['NIRMAL'],dict['JAGITYAL'])
    graph.add_edge('JAGITYAL','MANCHERIAL',dict['JAGITYAL'],dict['MANCHERIAL'])
    graph.add_edge('NIZAMABAD','KAMAREDDY',dict['NIZAMABAD'],dict['KAMAREDDY'])
    graph.add_edge('NIZAMABAD','SIRICILLA',dict['NIZAMABAD'],dict['SIRICILLA'])
    graph.add_edge('JAGITYAL','PEDDAPALLI',dict['JAGITYAL'],dict['PEDDAPALLI'])
    graph.add_edge('JAGITYAL','KARIMNAGAR',dict['JAGITYAL'],dict['KARIMNAGAR'])
    graph.add_edge('MANCHERIAL','PEDDAPALLI',dict['MANCHERIAL'],dict['PEDDAPALLI'])
    graph.add_edge('KAMAREDDY','SIRICILLA',dict['KAMAREDDY'],dict['SIRICILLA'])
    graph.add_edge('KAMAREDDY','SANGAREDDY',dict['KAMAREDDY'],dict['SANGAREDDY'])
    graph.add_edge('KAMAREDDY','MEDAK',dict['KAMAREDDY'],dict['MEDAK'])
    graph.add_edge('SIRICILLA','SIDDIPET',dict['SIRICILLA'],dict['SIDDIPET'])
    graph.add_edge('SIRICILLA','KARIMNAGAR',dict['SIRICILLA'],dict['KARIMNAGAR'])
    graph.add_edge('KARIMNAGAR','PEDDAPALLI',dict['KARIMNAGAR'],dict['PEDDAPALLI'])
    graph.add_edge('KARIMNAGAR','BHUPALPALLY',dict['KARIMNAGAR'],dict['BHUPALPALLY'])
    graph.add_edge('PEDDAPALLI','BHUPALPALLY',dict['PEDDAPALLI'],dict['BHUPALPALLY'])
    graph.add_edge('MEDAK','SANGAREDDY',dict['MEDAK'],dict['SANGAREDDY'])
    graph.add_edge('MEDAK','SIDDIPET',dict['MEDAK'],dict['SIDDIPET'])
    graph.add_edge('MEDAK','MEDCHAL',dict['MEDAK'],dict['MEDCHAL'])
    graph.add_edge('SIDDIPET','MEDCHAL',dict['SIDDIPET'],dict['MEDCHAL'])
    graph.add_edge('SIDDIPET','WARANGAL',dict['SIDDIPET'],dict['WARANGAL RURAL']+dict['WARANGAL URBAN'])
    graph.add_edge('SIDDIPET','JANGAON',dict['SIDDIPET'],dict['JANGAON'])
    graph.add_edge('SIDDIPET','YADADRI',dict['SIDDIPET'],dict['YADADRI'])
    graph.add_edge('SIDDIPET','KARIMNAGAR',dict['SIDDIPET'],dict['KARIMNAGAR'])
    graph.add_edge('WARANGAL','MULUGU',dict['WARANGAL RURAL']+dict['WARANGAL URBAN'],dict['MULUGU'])
    graph.add_edge('WARANGAL','BHUPALPALLY',dict['WARANGAL RURAL']+dict['WARANGAL URBAN'],dict['BHUPALPALLY'])
    graph.add_edge('WARANGAL','KARIMNAGAR',dict['WARANGAL RURAL']+dict['WARANGAL URBAN'],dict['KARIMNAGAR'])
    graph.add_edge('WARANGAL','JANGAON',dict['WARANGAL RURAL']+dict['WARANGAL URBAN'],dict['JANGAON'])
    graph.add_edge('WARANGAL','MAHABOOBNAGAR',dict['WARANGAL RURAL']+dict['WARANGAL URBAN'],dict['MAHABOOBNAGAR'])
    graph.add_edge('WARANGAL','MAHABUBABAD',dict['WARANGAL RURAL']+dict['WARANGAL URBAN'],dict['MAHABUBABAD'])
    graph.add_edge('SANGAREDDY','MEDCHAL',dict['SANGAREDDY'],dict['MEDCHAL'])
    graph.add_edge('SANGAREDDY','RANGAREDDY',dict['SANGAREDDY'],dict['RANGAREDDY'])
    graph.add_edge('SANGAREDDY','VIKARABAD',dict['SANGAREDDY'],dict['VIKARABAD'])
    graph.add_edge('MEDCHAL','YADADRI BHONGIR',dict['MEDCHAL'],dict['YADADRI'])
    graph.add_edge('MEDCHAL','GHMC',dict['MEDCHAL'],dict['GHMC'])
    graph.add_edge('MEDCHAL','RANGAREDDY',dict['MEDCHAL'],dict['RANGAREDDY'])
    graph.add_edge('YADADRI','JANGAON',dict['YADADRI'],dict['JANGAON'])
    graph.add_edge('YADADRI','RANGAREDDY',dict['YADADRI'],dict['RANGAREDDY'])
    graph.add_edge('YADADRI','NALGONDA',dict['YADADRI'],dict['NALGONDA'])
    graph.add_edge('YADADRI','SURYAPET',dict['YADADRI'],dict['SURYAPET'])
    graph.add_edge('JANGAON','MAHABUBABAD',dict['JANGAON'],dict['MAHABUBABAD'])
    graph.add_edge('JANGAON','SURYAPET',dict['JANGAON'],dict['SURYAPET'])
    graph.add_edge('MAHABUBABAD','KOTHAGUDEM',dict['MAHABUBABAD'],dict['KOTHAGUDEM'])
    graph.add_edge('RANGAREDDY','VIKARABAD',dict['RANGAREDDY'],dict['VIKARABAD'])
    graph.add_edge('RANGAREDDY','NARAYANPET',dict['RANGAREDDY'],dict['NARAYANPET'])
    graph.add_edge('RANGAREDDY','MAHABOOBNAGAR',dict['RANGAREDDY'],dict['MAHABOOBNAGAR'])
    graph.add_edge('RANGAREDDY','NAGARKURNOOL',dict['RANGAREDDY'],dict['NAGARKURNOOL'])
    graph.add_edge('RANGAREDDY','GHMC',dict['RANGAREDDY'],dict['GHMC'])
    graph.add_edge('RANGAREDDY','NALGONDA',dict['RANGAREDDY'],dict['NALGONDA'])
    graph.add_edge('VIKARABAD','NARAYANPET',dict['VIKARABAD'],dict['NARAYANPET'])
    graph.add_edge('VIKARABAD','MAHAB00BNAGAR',dict['VIKARABAD'],dict['MAHABOOBNAGAR'])
    graph.add_edge('NALGONDA','SURYAPET',dict['NALGONDA'],dict['SURYAPET'])
    graph.add_edge('NALGONDA','MAHABOOBNAGAR',dict['NALGONDA'],dict['MAHABOOBNAGAR'])
    graph.add_edge('NALGONDA','NAGARKURNOOL',dict['NALGONDA'],dict['NAGARKURNOOL'])
    graph.add_edge('SURYAPET','KHAMMAM',dict['SURYAPET'],dict['KHAMMAM'])
    graph.add_edge('KHAMMAM','MAHAB00BNAGAR',dict['KHAMMAM'],dict['MAHABOOBNAGAR'])
    graph.add_edge('KHAMMAM','KOTHAGUDEM',dict['KHAMMAM'],dict['KOTHAGUDEM'])
    graph.add_edge('KOTHAGUDEM','MAHAB00BNAGAR',dict['KOTHAGUDEM'],dict['MAHABOOBNAGAR'])
    graph.add_edge('NARAYANPET','MAHAB00BNAGAR',dict['NARAYANPET'],dict['MAHABOOBNAGAR'])
    graph.add_edge('NARAYANPET','GADWAL',dict['NARAYANPET'],dict['GADWAL'])
    graph.add_edge('MAHABOOBNAGAR','WANAPARTHY',dict['MAHABOOBNAGAR'],dict['WANAPARTHY'])
    graph.add_edge('MAHABOOBNAGAR','NAGARKURNOOL',dict['MAHABOOBNAGAR'],dict['NAGARKURNOOL'])
    graph.add_edge('GADWAL','WANAPARTHY',dict['GADWAL'],dict['WANAPARTHY'])
    graph.add_edge('WANAPARTHY','NAGARKURNOOL',dict['WANAPARTHY'],dict['NAGARKURNOOL'])  
    source=input(print('Enter starting point in CAPS:'))
    destination=input(print('Enter destination point in CAPS:'))
    Y=[]
     Y=shortest_path(graph,source,destination)
    print(Y)
```
